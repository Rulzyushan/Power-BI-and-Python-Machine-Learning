```python
#Load in the dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
dataset = pd.read_csv('HR_file.csv')
dataset.head()
```

```python
dataset.columns
```

```python
#Lets change catagories to numbers
le = LabelEncoder()
dataset['Departments '] = le.fit_transform(dataset['Departments '])
dataset['salary'] = le.fit_transform(dataset['salary'])
```

```python
dataset.head()
```

```python
y = dataset['Quit the Company']
features = ['Satisfaction Level', 'Last Evaluation', 'Number of Projects',
       'Monthly Hours', 'Total Time at the Company', 'Work Accidents', 'Promoted in Last 5 yrs', 'Departments ', 'salary',
       'Management']
x = dataset[features]
```

```python
#Lets scale the data
s = StandardScaler()
x = s.fit_transform(x)
```

```python
#Split and train the dataset
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
#Let the model predict results
log = LogisticRegression()
log.fit(x_train,y_train)
y_pred = log.predict(x)
y_prob = log.predict_proba(x)[:,1]
```

```python
#Lets add the column back to the dataset
dataset['prediction'] = y_pred
dataset['probabilities'] = y_prob
```
